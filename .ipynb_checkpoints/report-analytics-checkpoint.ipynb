```python
def get_prod(text):
    if 'E1' in text:
        return 'E1'
    if 'A1' in text:
        return 'A1'
    if 'A2' in text:
        return 'A2'
    if 'C1' in text:
        return 'C1'
    else:
        return 'None'
    
def cal_true_prob(x): 
    if x['top'] == True:
        return x['freq'] / (x['count'])
    else:
        return (x['count'] - x['freq'] )/ (x['count'])


level_3['prod'] = level_3['group'].apply(get_prod)
```

```python
sig_stat = level_3.groupby('prod')['greater_sig'].describe().reset_index()
sig_stat['true_prob'] = sig_stat.apply(cal_true_prob, axis=1).round(2)
sig_stat

sig_stat = level_3.groupby('prod')['y_score'].describe().reset_index()
sig_stat
```

```python
def age_range(num):
    if '18' in num:
        return '1-18'
    if  '27' in num:
        return '19-27'
    if '38' in num:
        return '28-38'
    if '45' in num :
        return '39-45'
    if '52' in num:
        return '46-52'
    if '53+' in num:
        return '53+'
    else:
        return 'None'

def annual_range(num):
    if '1-60,000' in num:
        return '1-60,000'
    if  '60,001-88,000' in num:
        return  '60,001-88,000' 
    if '88,001-130,0000' in num:
        return '88,001-130,0000'
    if '130,000-190,000' in num :
        return '130,000-190,000'
    if '190,001-300,0000' in num:
        return '190,001-300,0000'
    if '300,001+' in num:
        return '300,001+'
    else:
        return 'None'
    
def risk_level(num):
    if 'Very Risk' in num:
        return 'Very Risk'
    if  'Risk' in num:
        return  'Risk'
    if 'Normal' in num:
        return 'Normal'
    if 'Very Good' in num :
        return 'Very Good'
    if 'Good' in num:
        return 'Good'
    if 'Excellent' in num:
        return 'Excellent'
    else:
        return 'None'
    
def cust_gender(num):
    if 'F' in num:
        return 'F'
    if 'M' in num:
        return 'M'
    else:
        return 'None'
    
def cus_marital_status(num):
    if 'NA' in num:
        return '00 NA'
    if '01 Single' in num:
        return '01 Single'
    if '02 Maried' in num:
        return '02 Maried'
    if '04 Divorce' in num:
        return '04 Divorce'
    if '03 Widow' in num:
        return '03 Widow'
    else:
        return 'None'

level_3['age_range'] = level_3['group'].apply(age_range)
level_3['annual_range'] = level_3['group'].apply(annual_range)
level_3['risk_level'] = level_3['group'].apply(risk_level)
level_3['cust_gender'] = level_3['group'].apply(cust_gender)
level_3['cus_marital_status'] = level_3['group'].apply(cus_marital_status)
```

```python
sig_stat = level_3.groupby('age_range')['greater_sig'].describe().reset_index()
sig_stat['true_prob'] = sig_stat.apply(cal_true_prob, axis=1).round(2)
sig_stat

sig_stat = level_3.groupby('age_range')['y_score'].describe().reset_index()
sig_stat
```
