```python
df = pd.read_csv('data/testReport_level_all.csv')

df['y_2018'] = df['y_2018'].round(2)
df['y_2017'] = df['y_2017'].round(2)
df['y_score'] = df['y_score'].round(2)
df['premium_2018_Mean'] = df['premium_2018_Mean'].round(0)
df['premium_2017_Mean'] = df['premium_2017_Mean'].round(0)
df['greater_p_value'] = df['greater_p_value'].round(5)
```

```python
def get_prod(text):
    if 'E1' in text:
        return 'E1'
    if 'A1' in text:
        return 'A1'
    if 'A2' in text:
        return 'A2'
    if 'C1' in text:
        return 'C1'
    else:
        return 'None'

level_3['prod'] = level_3['group'].apply(get_prod)
```

```python
sig_stat = level_3.groupby('prod')['greater_sig'].describe().reset_index()
```
